```python
uniform_data= stats.uniform.rvs( size= 10000,  #generate 10000 numbers
                                 loc=0,       #From 0
                                 scale=10)    #To 10
```

```python
pd.DataFrame(uniform_data).plot(kind = "density",  #plot the values
                                figsize= (10,9),
                                xlim=(-1,11));
```

```python
#create a quantile (percentile) comes under cumulative distribution function
# It's not similar or equal to percentage
#25 percentile mean 25 percent scored marks less than mass u got
#cdf is cumulative data function(mean)

stats.uniform.cdf(x=2,        #cutoff value (quantile)
                  loc=0,      #distribution start
                  scale=10)  #DISTRIBUTION END
```

```python
#USING stats.distribution.ppf is inverse of cdf :It returns x axis cutopp value (quntile) associated with a given probability.
stats.uniform.ppf(q = 0.4,      #probability cutoff
                  loc = 0,      #distribution cutoff
                  scale=10)  #Distribution end
```

```python
combined_table = pd.merge(table01,         #First table
                          table02,         #Second table
                          how= "inner",    #Merge method
                          on= "P_ID")     #column to join on
combined_table
```

```python
#while analysing the cobined table we cn see that some of them are missing but we want all values to be merged

#Use merge to perform other types of joins ,such as left join and outer joins

#A left join keep all values in the first (left) of the dataframe

left_join = pd.merge(table01,         #First table
                          table02,         #Second table
                          how= "left",    #Merge method
                          on= "P_ID")     #column to join on
left_join

#NaN=Not a number
```

```python
#A right join keeps all values in the second(right) dataframe

right_join = pd.merge(table01,         #First table
                          table02,         #Second table
                          how= "right",    #Merge method
                          on= "P_ID")     #column to join on
right_join
```

```python
#An outer join keeps all values in both dataframes

outer_join = pd.merge(table01,         #First table
                          table02,         #Second table
                          how= "outer",    #Merge method
                          on= "P_ID")     #column to join on
outer_join
```
